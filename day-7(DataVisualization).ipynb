```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x = np.linspace(0.0,5.0,100)
y = np.sin(x)

# print(x)
# print(y)

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot Ex')
plt.grid(True)
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([3,5,7,2,1])

plt.scatter(x,y, c='red', marker='o', edgecolors='black')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot Ex')
plt.grid(True)
plt.show()
```

```python
cat = ['A','B','C','D']
valu = [20,35,10,15]


plt.bar(cat,valu, color = ['red','green','blue','pink'])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Example')
plt.xticks(rotation=180)
plt.show()
```

```python
x= np.random.normal(loc=50, scale=10, size=1000)

plt.hist(x,bins =20, edgecolor = 'black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.grid(True)
plt.show()
```

```python
data = np.random.rand(11,11)
print(data)

plt.imshow(data, cmap='hot')
plt.colorbar(label='value')
```

```python
fruits = np.array(['Apple','Orange','Banana'])
count = np.array([20,30,10])

plt.pie(count,labels = fruits , autopct = '%1.1f%%')
plt.title('pie')
plt.show()
```

```python
sco = {'class A' : [60,64,68,70,75],
        'class B' : [40,44,48,50,55],
        'class C' : [20,24,28,30,35]}

plt.boxplot(sco.values(), labels=sco.keys())
plt.title('boxplot')
plt.show()
```

```python
sco = {'Temperature' : [60,64,68,70,75],
        'Humadity' : [40,44,48,50,55],
        'Growth' : [5,6,2,8,3]}

data = pd.DataFrame(sco)

sns.pairplot(data)
plt.show()
```
